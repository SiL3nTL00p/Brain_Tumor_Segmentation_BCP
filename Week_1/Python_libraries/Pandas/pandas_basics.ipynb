```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

data = pd.read_csv("videogamesales.csv")
print(data)
data["global_sales"] = data["NA_Sales"]+data["EU_Sales"]+data["JP_Sales"]+data["Other_Sales"]
data = data.sort_values(by="global_sales",ascending=False)

df_genre = data.groupby("Genre")["global_sales"].sum()
print(df_genre)

plt.figure(figsize=(8, 8))
plt.bar(df_genre.index,df_genre.values,width=0.8,color="lightcoral",edgecolor="black")
plt.xlabel("Genre types")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.show()

df_gta = data[data["Name"].str.contains("Grand Theft Auto", na=False)]
print(df_gta[["Name", "Platform", "Year", "EU_Sales", "JP_Sales"]])


total_na = df_gta["NA_Sales"].sum()
total_eu = df_gta["EU_Sales"].sum()
total_jp = df_gta["JP_Sales"].sum()
total_other = df_gta["Other_Sales"].sum()

sales_data = [total_na, total_eu, total_jp, total_other]
labels = ["North America", "Europe", "Japan", "Other"]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

plt.figure(figsize=(8, 8))
plt.pie(sales_data, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=(0.05, 0, 0, 0))
plt.title("Total Grand Theft Auto Sales by Region")
plt.show()
```
